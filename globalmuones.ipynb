```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

archivo = pd.read_csv("https://goo.gl/sp8gYx")
archivo.head()
```

```python
#datos limpios
Datos_limpios = archivo[archivo['type2'] == 'G']
Datos_limpios.head()
```

```python
P_t_1_l = np.sqrt(Datos_limpios["px1"]**2 + Datos_limpios["py1"]**2) #Momento transversal de la particula 1 limpio
P_t_2_l = np.sqrt(Datos_limpios["px2"]**2 + Datos_limpios["py2"]**2) #Momento transversal de la particula 2 limpio
```

### Estas líneas de código  se usarán para graficar con matplotlib.pyplot


```python

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.hist(P_t_2_l,100) # Histograma del momento transversal reconstruido 
plt.title("Reconstruccion del momento transversal limpio",size=16)
plt.xlabel(r"$P_{T2}$", size=20)
plt.ylabel("Eventos",size=12)
plt.xlim(0,300)
plt.yticks(size=14)
plt.xticks(np.linspace(0,300,7),size=14)
plt.grid()
plt.subplot(1,2,2) 
plt.title("Momento transversal de los datos limpio",size=16)
plt.hist(Datos_limpios["pt2"],1000) #Momento transversal proporcionado en el Sample1
plt.xlabel(r"$P_{T2}$", size=20)
plt.ylabel("Eventos",size=12)
plt.xlim(0,300)
plt.yticks(size=14)
plt.xticks(np.linspace(0,300,7),size=14)
plt.grid()
plt.savefig("pt2_lim.png")

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.hist(P_t_1_l,2000) # Histograma del momento transversal reconstruido 
plt.title("Reconstruccion del momento transversal limpio",size=16)
plt.xlabel(r"$P_{T1}$", size=20)
plt.ylabel("Eventos",size=12)
plt.xlim(0,300)
plt.yticks(size=14)
plt.xticks(np.linspace(0,300,7),size=14)
plt.grid()
plt.subplot(1,2,2) 
plt.title("Momento transversal de los datos limpio",size=16)
plt.hist(Datos_limpios["pt1"],2000) #Momento transversal proporcionado en el Sample1
plt.xlabel(r"$P_{T1}$", size=20)
plt.ylabel("Eventos",size=12)
plt.xlim(0,300)
plt.yticks(size=14)
plt.xticks(np.linspace(0,300,7),size=14)
plt.grid()
plt.savefig("pt1_lim.png")
```

##Lineas de codigo para graficar la masa invariante con matplotlib

```python
E_2_a = (Datos_limpios["E1"]+Datos_limpios["E2"])**2 #Creo los datos para la energia al cuadrado
P_2_a = ((Datos_limpios["px1"]+Datos_limpios["px2"])**2+(Datos_limpios["py1"]+Datos_limpios["py2"])**2+(Datos_limpios["pz1"]+Datos_limpios["pz2"])**2) #El momento al cuadrado
M2_a = E_2_a-P_2_a #Vector con la masa invariante al cuadrado, es posible desde aqui sacar la raiz cuadrada

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.hist(np.sqrt(M2_a),1000) #Histograma de la masa reconstruida
plt.xlim(0,100)
plt.title("Masa reconstruida",size=16)
plt.xlabel("Masa [GeV]",size=16)
plt.ylabel("Eventos",size=12)
plt.grid()
plt.yticks(size=14)
plt.xticks(size=14)
plt.xticks(np.linspace(0,200,11),size=14)
plt.subplot(1,2,2)
plt.hist(Datos_limpios["M"],1000)
plt.xlim(0,100)
plt.title("Masa de los datos",size=16) #Histograma de la masa dada en el sample1
plt.xlabel("Masa [GeV]",size=16)
plt.ylabel("Eventos",size=12)
plt.grid()
plt.yticks(size=14)
plt.xticks(np.linspace(0,200,11),size=14)
plt.savefig("Masa_lim.png")
```
